```python
import numpy as np # To working with these arrays. 
import pandas as pd # to make working with “relational” or “labeled” data  easy 
import matplotlib.pyplot as plt #To creates a figure, creates a plotting area in a figure, plots some lines in a plotting area,
```

```python
%matplotlib inline
```

```python
#Reading  dataset set  
ds = pd.read_csv('cluster_table.csv')                     
ds.head()
```

```python
# Check for null values
ds.info()
```

```python
ds.describe() #To know mathematical values like min, max,avg of data in each coloumn
```

### This data has no missing values and the types are of float64 as displayed above

```python
# creating a data frame with cluster data.
cd = ds[['X', 'Y', 'Z']]                       
```

```python
#  new data frame
cd.head()
```

```python
# extracting each colom separatetly 
x = ds.X.values
y = ds.Y.values
z = ds.Z.values

# tranform to a 1-dim vector
values = np.array([x,y,z]).transpose()      
```

```python
# Standardize features by removing the mean and scaling to unit variance

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(values)
sc_val = scaler.transform(values)  
sc_val
```

```python
# DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a popular unsupervised learning method utilized in model building and machine learning algorithms 
from sklearn.cluster import DBSCAN
from sklearn import metrics


dbs = DBSCAN()             

# fit the model on the data also known as trainning the data
db = dbs.fit(scaled_val)

# get all cluster labels  
cl_labels = db.labels_

# Increment the cluster number by 1 
cl_labels=cluster_labels+1
cluster_labels
```

```python
# import the plotting library fot 3d plots
from mpl_toolkits.mplot3d import Axes3D

# create a fig and add to the axes
fig = plt.figure(figsize=[15,8])
ax = fig.add_subplot(111, projection='3d')

ax.scatter(sc_val[:,0],sc_val[:,1],sc_val[:,2], c=cl_labels)
```

```python
# import all model performance measurement libraries
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score

# set predicted values to cluster labels
pred_val = cluster_labels

true_val = ds['Cluster label'].values.astype(int)
```

```python
# display the clustering using the original cluster labels
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=[15,10])
ax = fig.add_subplot(111, projection='3d')
ax.scatter(ds.X, ds.Y, ds.Z, c=ds['Cluster label'])
```

```python
print(confusion_matrix(true_val, pred_val))
```

#### It can be deduced from the confusion matrix that:
- There is noise in the clustering and this is a typical characteristic of the DBSCAN algorithm.
- along the diagonal we have all true positives, i.e the clusters data points that were correctly predicted, 
- this accounts sfor the high f1 score.

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(true_val, pred_val))
```
